# Signals

The most common types of signals found includes the Sinusoid wave forms. They have the following general form: <br>
$y(t) = Asin(\omega t+\phi) = Asin(2\pi ft+\phi)$ <br>
where: <br>
$A$ is the amplitude <br>
$t$ is the real independent variable (usually time) <br>
$\omega$ is the angular frequency (in radians per second) <br>
$f$, ordinary frequency (in Hz), the number of oscillations that occur each second of time <br>
$\phi$, phase, specifies (in radians) where in the cycle the oscillation is at t=0.

Looking into $\phi$, the phase shift. If $\phi$ is non-zero, the resultant waveform would be shifted in time by $\frac{\phi}{\omega}$ seconds. <br>
If $\phi < 0$, the waveform is delayed (shifted to the right). <br>
If $\phi > 0$, the waveform is advanced (shifted to the left).

```python
# Sample Sinusoid Wave
def plot_sinusoid(amplitude=1.0, frequency=1.0, phase=0.0, axis_limit=2):
    t = np.linspace(-axis_limit * np.pi , axis_limit * np.pi, 10000)
    y = amplitude * np.sin( 2 * np.pi * frequency * t + phase)

    plt.figure(figsize=(10, 4))
    plt.plot(t, y, label=f'{amplitude} sin(2* pi * {frequency}t + {phase})')
    plt.axhline(0, color='black', linestyle='--', linewidth=1)
    plt.axvline(0, color='black', linestyle='--', linewidth=1)
    plt.title('Sinusoidal Wave')
    plt.xlabel('t')
    plt.ylabel('y')
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
sliders = {
    "amplitude": FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0),
    "frequency": FloatSlider(min=0.1, max=5.0, step=0.1, value=1.0),
    "phase": FloatSlider(min=-2*np.pi, max=2 * np.pi, step=0.1, value=0.0),
}

ui = interactive(plot_sinusoid, **sliders)
display(ui)
```

### Relationship between Sinusoid and Cosine Signal

Cosine is equivalent to Sinusoid by phase shifting ($\phi$) a sinusoid signal by $\frac{\pi}{2}$

```python
# Sinusoid Wave
plot_sinusoid(1, 1, 0, 1)
```

```python
# Cosine Wave
plot_sinusoid(1, 1, np.pi/2,1)
```
